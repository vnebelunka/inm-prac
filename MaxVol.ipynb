```python
m = 100
r = 5

A = np.random.randn(m, r)
A
```

```python
def Sherman_Morrison_inv(A_inv, u, v):
    den = 1 + (v.T @ A_inv @ u)
    return A_inv - np.outer(A_inv @ u,  v.T @ A_inv) / den
```

```python
def MaxVol(A, eps=1e-2, maxit=100):
    m, r = A.shape
    set_i = list(range(0, r))
    cur_vol = abs(np.linalg.det(A[set_i, :]))
    
    print("it: 0: {}".format(cur_vol))
    C = A[list(set_i), :]
    C_inv = np.linalg.inv(C)    
    for it in range(1, maxit):
        A_temp = np.abs(A @ C_inv)
        i, k = np.unravel_index(A_temp.argmax(), A_temp.shape)
        a_i = A[i, :]
        a_k = A[k, :]
        
        print("it: {}: {}".format(it, cur_vol))        
        if A_temp[i, k] <= 1 + eps:
            break
        set_i[k] = i
        cur_vol *= A_temp[i, k]
        
        # C^-1 := (C - e_k(a_i - a_k))^-1
        C_inv = Sherman_Morrison_inv(C_inv, -np.eye(1, r, k=k).T, (C[k, :] - a_i)[None, :].T)
    return set_i
```

```python
argvol = MaxVol(A)
abs(np.linalg.det(A[argvol, :]))
```
